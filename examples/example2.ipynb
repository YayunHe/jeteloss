# Example usage of the jeteloss extractor

This example is built for jupyter-notebook

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from jeteloss import PythiaPP, RAA2Eloss
from scipy.special import gamma

sns.set_context('poster')
sns.set_style('ticks')
```

```python
pp_x, pp_y = PythiaPP(2760)
```

```python
eloss = RAA2Eloss("RAA_2760.txt", pp_x, pp_y)
```

```python
eloss.train(steps=400000, burn=200000, thin=10)
```

```python
# the lengh of the eloss.trace('a')[...]
len(eloss.a)
```

```python
eloss.plot_mean_ptloss(jet_pt=np.linspace(1, 400, 100), jupyter=True)
```

```python
eloss.plot_pt_loss_dist(jupyter=True)
```

```python
eloss.plot_RAA(jupyter=True)
```
